```python
import uproot
import pandas as pd

root_filename = 'OptSim_LUT_ArgonCube2x2_32.root'
root_file = uproot.open(root_filename)
root_file.keys()
```

```python
tree = root_file['PhotonLibraryData']
branch = tree.arrays()
#branches = tree.arrays(namedecode='utf-8')
branch.keys()
```

```python
#df = tree.pandas.df(flatten=False)
df = tree.pandas.df()
df.head()
```

```python
import plotly.express as px
import numpy as np

fig = px.scatter(
    df,
    title='OptSimLUT',
    x='OpChannel',
    y='YPos',
    color='VolIdx',
    color_continuous_scale=['darkblue','red'],
    template='seaborn')
fig.update_layout(yaxis_type='linear', yaxis_title='YPos')
fig.show()
```

```python
import plotly.graph_objects as go
fig = px.density_heatmap(
    df,
    title='OptSimLUT',
    x='OpChannel',
    y='Voxel',
    z=np.log(df['Visibility']*100),
    nbinsx=48,
    nbinsy=32,
    #color=np.log(df['Visibility']*100),
    color_continuous_scale=['darkblue','red','gold'],
    template='seaborn')
fig.update_layout(yaxis_type='linear', yaxis_title='Voxel', coloraxis_colorbar_title='Log Visibility [%]')
fig.show()
```
